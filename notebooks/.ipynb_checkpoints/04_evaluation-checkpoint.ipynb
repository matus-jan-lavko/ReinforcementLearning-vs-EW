# Load results and Data

```python
#load actions and returns
weights_RL = {mod : pd.read_csv(os.path.join(RESULTS_PATH,DATASET,mod,str.lower(mod) + '_actions.csv'),index_col=0,parse_dates=True) for mod in MODELS}
r_full = pd.read_csv(os.path.join(DATA_PATH, DATASET + '__PX_LAST.csv'),index_col=0,parse_dates=True)
test_index = weights_RL['A2C'].index

#load backtests
backtests = {window : pd.read_csv(os.path.join(RESULTS_PATH,DATASET,str(window) + '.csv'), 
                                 index_col = 0, parse_dates = True) for window in REBALANCING_WINDOW.keys()}

#compute cumulative returns
backtests_cuml = {window : (1 + backtests[window]).cumprod() for window in REBALANCING_WINDOW.keys()}

#reindex to 1
for window in backtests_cuml.keys():
    backtests_cuml[window].iloc[0, :] = 1


#load risk-free and auxiliary data
r = r_full.loc[test_index, weights_RL['A2C'].columns]
rf = pd.read_csv(os.path.join(DATA_PATH, RISK_FREE), index_col = 0, parse_dates = True).loc[test_index]
rf /= 100 #is in percentages
rf = rf.iloc[:,0] #take the 10Y US T-Bill

db_factors = pd.read_excel(os.path.join(DATA_PATH, FACTORS), index_col = 0, parse_dates = True)
```

# Evaluation

```python
def plot_on_axis(ax, df, title):
    ax.plot(df['EW'], color = 'red',linewidth=1)
    for mod in MODELS:
        ax.plot(df[mod],linewidth = 1,linestyle='--')
    ax.set_title(title)
```

```python
fig, axes = plt.subplots(ncols = 2, nrows = 2, constrained_layout = True, figsize=(18,18))
plot_on_axis(axes[0][0], backtests_cuml['Weekly'],'Weekly')
plot_on_axis(axes[0][1], backtests_cuml['Monthly'],'Monthly')
plot_on_axis(axes[1][0], backtests_cuml['Quarterly'],'Quarterly')
plot_on_axis(axes[1][1], backtests_cuml['Semi-annually'],'Semi-annually')
fig.legend(['EW'] + MODELS, loc = 5, bbox_to_anchor=(0.554, 1.02), ncol=2, fancybox = True, shadow = True)
```

```python
def outperform_mask(backtests, summary_stats, metric: str, side = 'right'):
    if side == 'right':
        subset = summary_stats.loc[:, summary_stats.loc[metric] > summary_stats.loc[metric, 'EW']].columns.to_list()
        
    elif side == 'left':
        subset = summary_stats.loc[:, summary_stats.loc[metric] < summary_stats.loc[metric, 'EW']].columns.to_list()
    else:
        raise ValueError
    subset = pd.DataFrame(backtests, columns = subset)
    ew = pd.DataFrame(backtests, columns = ['EW'])
    return subset, ew
```

```python
def perf_counter(function):
    def wrapper(*args, **kwargs):
        tic = time.perf_counter()
        rtrn = function(*args, **kwargs)
        toc = time.perf_counter()
        print(f'{function.__name__!r} in {toc - tic :.4f} seconds')
        return rtrn
    return wrapper
```

```python
@perf_counter
def display_gif(model):
    return Image(os.path.join(RESULTS_PATH, DATASET, model, model + '_weights.gif'))

```

```python
@perf_counter
def display_image(model):
    try:
        return Image(os.path.join(RESULTS_PATH, DATASET, model, 'tb_train.jpg'))
    except FileNotFoundError:
        print('No Tensorboard Picture!')
        
```
